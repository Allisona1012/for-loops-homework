```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 10 
num2 = 5

if num1 > num2:
    print ("the conditon is true")
print ('finished')
```

```python
if 5 > 3 and 15 < 5:
    print ("Both are true")
    
if 5>1 or 6 <3:
    print ("At least one is true")
```

##### Continue Statement

```python
# will continue to next iteration
```

```python
print('loop started')

for i in range(20):
    if i == 5:
        continue
    print(i)

print("loop ended")
```

##### Break Statement

```python
# will break out of current loop
```

```python
print('loop started')

for i in range(20):
    if i == 5:
        break
    print(i)

print("loop ended")
```

##### Pass Statement

```python
# mostly used as a placeholder to fill codeblock, and will continue on same iteration
```

```python
print('loop started')

for i in range(20):
    if i == 5:
       pass
    print(i)

print("loop ended")
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didnt
# An else after a for runs if the for didnt break

# a prime number must be greater than 1
for num in range(100):
    if num >1:
        for i in range (2, num):
            if num % i == 0:
                break
        else:
            print (num)

            
```
